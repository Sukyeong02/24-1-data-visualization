<h1>문제 1</h1>
bank.csv 파일에서 직업(job) 컬럼에 기록된 9개의 직업 유형별 은행 잔고(balance) 데이터를 상자 그림(Box Plot)으로 시각화하세요.<br>
<br>9가지 직업 종류<br>
1. admin.<br>
2. blue-collar<br>
3. entrepreneur<br>
4. housemaid<br>
5. management<br>
6. self-employed<br>
7. services<br>
8 student<br>
9. technician<br>

<h4>제약사항</h4>
* 그래프의 가로 및 세로 길이는 각각 15, 3인치로 지정하세요.<br>
* 그래프의 제목, 세로 축의 이름, 격자 등 세부 사항은 셀 아래 그래프를 참고하여 설정하세요.<br>
* NumPy 또는 pandas 라이브러리에서 제공하는 함수 또는 클래스 등을 사용할 경우 감점합니다.<br>
* 우리는 아직 NumPy, pandas 라이브러리를 공부하지 않았습니다.<br>

```python
import csv
import matplotlib.pyplot as plt

f = open("bank.csv", 'r', encoding="cp949")
data = csv.reader(f)

# 아래부터 솜솜이가 작성해야 할 부분
job = [[],[],[],[],[],[],[],[],[]]
job_name = ['admin.','blue-collar','entrepreneur','housemaid','management','self-employed','services','student','technician']

for row in data:
    for i in range(len(job_name)):
        if row[0] == job_name[i]:
            job[i].append(int(row[2]))

plt.figure(figsize=(15,3))
plt.boxplot(job, showfliers=False)
plt.title('Boxplot of Balance')
plt.xticks(range(1,10), job_name)
plt.ylabel('balance')
plt.grid(True)
plt.show()

# 수고하셨습니다.
f.close()
```

<h1>문제 2</h1>
bank.csv 파일에서 직업(job) 컬럼에 기록된 직업 유형 중 기술자(technician)와 학생(student)의 은행 잔고(balance) 및 예치기간(duration) 데이터를 산점도(scatter plot)으로 시각화하세요.<br>
이때, 가로축과 세로축을 각각 예치기간, 은행 잔고로 지정하고, 기술자는 빨간색 그리고 학생은 파란색으로 표시하세요.

<h4>제약사항</h4>
* 그래프의 가로 및 세로 길이는 기본값으로 사용하세요.<br>
* 그래프의 제목, 가로세로 축의 이름, 격자, 범례 등 세부 사항은 셀 아래 그래프를 참고하여 설정하세요.<br>
* NumPy 또는 pandas 라이브러리에서 제공하는 함수 또는 클래스 등을 사용할 경우 감점합니다.<br>
* 우리는 아직 NumPy, pandas 라이브러리를 공부하지 않았습니다.<br>

```python
import csv
import matplotlib.pyplot as plt

f = open("bank.csv", 'r', encoding="cp949")
data = csv.reader(f)

# 아래부터 솜솜이가 작성해야 할 부분
technician_balance = []
technician_duration = []
student_balance = []
student_duration = []

for row in data:
    if row[0] == 'technician':
        technician_balance.append(int(row[2]))
        technician_duration.append(int(row[3]))
    elif row[0] == 'student':
        student_balance.append(int(row[2]))
        student_duration.append(int(row[3]))

plt.scatter(technician_duration, technician_balance, color='red', label='technician')
plt.scatter(student_duration, student_balance, color='blue', label='student')
plt.xlabel('duration')
plt.ylabel('balance')
plt.legend()
plt.grid(True)
plt.show()

# 수고하셨습니다.
f.close()
```
